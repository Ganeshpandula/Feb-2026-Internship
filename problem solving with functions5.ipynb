```python
#Social Media – Post Engagement Analyzer
def analyze_likes(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
analyze_likes([300, 300, 250, 200])
```

#Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall_checker([70, 75, 68, 75], 70)
```

#System Monitoring – Application Health Checker

#E-Learning – Student Attendance Eligibility 

```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    for status in attendance_list:
        if status == 1:
            attended += 1
    percentage = (attended / total_classes) * 100
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", result)
# Ex (1 = Present, 0 = Absent)
attendance_checker([1,1,1,0,1])
```

#Smart Electricity Bill Analyzer


```python
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = units*5
    else:
        bill = units* 7
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    print("Total Bill:", bill)
    print("Usage Level:", usage)
electricity_bill(250)
```
